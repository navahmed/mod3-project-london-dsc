```python
# Create an empty dictionary ordinal_enc_dict
ordinal_enc_dict = {}

for col_name in terry_prac:
    # Create Ordinal encoder for col
    ordinal_enc_dict[col_name] = OrdinalEncoder()
    col = terry_prac[col_name]
    
    # Select non-null values of col
    col_not_null = col[col.notnull()]
    reshaped_vals = col_not_null.values.reshape(-1, 1)
    encoded_vals = ordinal_enc_dict[col_name].fit_transform(reshaped_vals)
    
    # Store the values to non-null values of the column in users
    terry_prac.loc[col.notnull(), col_name] = np.squeeze(encoded_vals)
```

```python
terry_prac.head()
```

```python
from sklearn.impute import KNNImputer
from fancyimpute import KNN
```

```python
# Create KNN imputer
KNN_imputer = KNN()

# Impute and round the users DataFrame
terry_prac.iloc[:, :] = np.round(KNN_imputer.fit_transform(terry_prac))

# Loop over the column names in users
for col_name in terry_prac:
    
    # Reshape the data
    reshaped = terry_prac[col_name].values.reshape(-1, 1)
    
    # Perform inverse transform of the ordinally encoded columns
    terry_prac[col_name] = ordinal_enc_dict[col_name].inverse_transform(reshaped)
```

Further Data Cleaning

We will assign the null values under '-' with the mode of the column for 'Frisk_Flag'. We will then assign a binary value to his column, with Y = 1 and N = 0.

```python
terry_clean['Frisk_Flag'].value_counts(dropna=False)
```

```python
terry_clean['Frisk_Flag'].replace('-','N', inplace=True)
```

```python
terry_clean['Frisk_Flag'].value_counts()
```

```python
terry_clean.loc[(terry_clean['Frisk_Flag']=='N'),'Frisk_Flag'] = 0
terry_clean.loc[(terry_clean['Frisk_Flag']=='Y'),'Frisk_Flag'] = 1
```

```python
terry_clean['Frisk_Flag'].value_counts()
```

```python
terry_clean.columns
```

```python
terry_clean['Year'] = terry_clean['Reported_Date'].apply(lambda x:x[:4])
terry_clean['Month']=terry_clean['Reported_Date'].apply(lambda x: x[5:7])
```

We will work out the Officer Age, by taking the difference between Officer YOB and Reported Date in order to determine the officer's age at the time of the incident.

```python
terry_clean['Officer_YOB'].value_counts()
```

```python
import numpy as np
```

```python
terry_clean['Year'] = terry_clean['Year'].astype(np.int64)
```

```python
terry_clean['Officer_Age'] = terry_clean['Year'] - terry_clean['Officer_YOB']
```

Will will now club age groups into decades as follows:

20-29

30-39

40-49

50-59

60+

```python
terry_clean.loc[(terry_clean['Officer_Age'] <= 29), 'Officer_Age_Group'] = '20-29'
terry_clean.loc[((terry_clean['Officer_Age'] >= 30) & (terry_clean['Officer_Age'] <= 39)), 'Officer_Age_Group'] = '30-39'
terry_clean.loc[((terry_clean['Officer_Age'] >= 40) & (terry_clean['Officer_Age'] <= 49)), 'Officer_Age_Group'] = '40-49'
terry_clean.loc[((terry_clean['Officer_Age'] >= 50) & (terry_clean['Officer_Age'] <= 59)), 'Officer_Age_Group'] = '50-59'
terry_clean.loc[(terry_clean['Officer_Age'] >= 60), 'Officer_Age_Group'] = '60+'
```
